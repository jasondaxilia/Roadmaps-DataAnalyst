```python
import csv
import pandas as pd

df = pd.read_csv('./data/salesdaily.csv')
print(df.head())
```

```python
# 1. What are the total sales quantities for each drug category (ATC code)?
print("1. Total sales quantities for each drug category (ATC code):")
non_category = ['datum','Year','Month','Weekday Name','Hour']
print(df.drop(columns=non_category).sum())
```

```python
# 2. Which individual drug brands have the highest total sales?
print("2. Individual drug brands that having the highest total sales is")
print("Which individual drug brands have the highest total sales?")
non_category = ['datum','Year','Month','Weekday Name','Hour']
total_sales = df.drop(columns=non_category).sum()
print(total_sales.sort_values(ascending=False).head(1))
```

```python
# 3. Which three drugs have the highest sales in January 2015, July 2016, September 2017.
print("3. Three drugs that have the highest sales in January 2015, July 2016, September 2017 is")
jan2015 = df[(df['Year'] == 2015) & (df['Month'] == 1)]
july2016 = df[(df['Year'] == 2016) & (df['Month'] == 7)]
sep2017 = df[(df['Year'] == 2017) & (df['Month'] == 9)]

print(f"January 2015:\n{jan2015.iloc[:, 1:9].sum().sort_values(ascending=False).head(3)}\n")
print(f"July 2016:\n{july2016.iloc[:, 1:9].sum().sort_values(ascending=False).head(3)}\n")
print(f"September 2017:\n{sep2017.iloc[:, 1:9].sum().sort_values(ascending=False).head(3)}\n")
```

```python
# 4. Which drug has sold the most often in 2017?
print("4. Which drug has sold the most often in 2017?")
y2017 = df[df['Year'] == 2017].iloc[:, 1:9]
print((y2017 > 0).sum().sort_values(ascending=False).head(1))
```

```python
# 5. Which drug category has the highest average daily sales?
print("5. Drug category with the highest average daily sales")
print(df.iloc[:, 1:9].mean().sort_values(ascending=False).head(1))
```

```python
# 6. Are respiratory drugs (R03) sold more during specific months?
print(df.groupby('Month')['R03'].sum().sort_values(ascending=False).head(1))
```
